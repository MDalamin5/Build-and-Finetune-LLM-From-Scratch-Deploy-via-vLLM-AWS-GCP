## **Finding Tensor data types**

```python
my_tensor = torch.tensor(
    [
        [1, 2, 3],
        [2, 3, 4]
    ]
)
my_tensor
```

```python
my_tensor.dtype
```

```python
my_tensor_int_16 = torch.tensor(
    [
        [1, 2, 3],
        [2, 3, 4]
    ],
    dtype = torch.int16
)
my_tensor_int_16.dtype
```

## Transfer one datatype to desier data type `.to` 

```python
flot_data = my_tensor_int_16.to(torch.float64)
flot_data
```
