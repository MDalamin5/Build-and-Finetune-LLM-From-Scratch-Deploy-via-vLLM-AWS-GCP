```python
from sklearn.datasets import make_classification
import torch
```

```python
X, y = make_classification(
    n_samples=10,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    random_state=42
)
X
```

```python
y
```

```python
X = torch.tensor(X, dtype=torch.float64)
y = torch.tensor(y, dtype=torch.long)
```

```python
X, y
```

## **Start of job Dataset and Dataloader class**

```python
X.shape[0]
```

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class MyCustomDatasetClass(Dataset):

    def __init__(self, features, labels):
        ## data loading is here.
        self.features = features
        self.labels = labels

    def __len__(self):
        return self.features.shape[0]

    def __getitem__(self, index):
        ## apply data transforms here.
        return self.features[index], self.labels[index]
```

```python
dataset = MyCustomDatasetClass(X, y)
```

```python
len(dataset)
```

```python
dataset[0]
```

## **Dataloader using Dataset class**

```python
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)
```

```python
for batch_features, batch_labels in dataloader:
    print(f"Batch Features: {batch_features} and labels {batch_labels} ")
    print("="*75)
```

## **Data Parallizations**
- using `worker` can be parralizations

```python
dataloader = DataLoader(dataset, batch_size=2, shuffle=True, num_workers=2, pin_memory=True, drop_last=False)
```

```python
for batch_features, batch_labels in dataloader:
    print(f"Batch Features: {batch_features} and labels {batch_labels} ")
    print("="*75)
```
