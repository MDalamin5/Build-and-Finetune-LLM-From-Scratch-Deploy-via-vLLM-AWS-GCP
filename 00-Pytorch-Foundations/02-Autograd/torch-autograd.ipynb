## **Build a simple perceptorn**

```python
x = torch.tensor(6.7)
y = torch.tensor(0.)
```

```python
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0., requires_grad=True)
```

```python
z = w*x + b
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
import torch.nn.functional as F
loss = F.binary_cross_entropy(y_pred, y)
loss
```

```python
loss.backward()
```

```python
w.grad, b.grad
```

## **Clening Grad**
- if we do not cleaning the grad then next time when we pass the forward pass then its auto add the previoud grad value so have to do `zero_grad or clear the grad`.

```python
x = torch.tensor(3., requires_grad=True)
x
```

```python
y = x**2
y
```

```python
y.backward()
```

```python
x.grad
```

## **Cleaning the grad using `Zero grad`**

```python
x = torch.tensor(2., requires_grad=True)
x
```

```python
y = x**2
y
```

```python
y.backward()
x.grad
```

```python
x.grad.zero_()
```

## **How to Disable Gradent Tracking**
- requires_grad = False
- detach()
- torch.no_grad()

```python
x.requires_grad_(False)
```

```python
y = x**2
```

```python
y
```

```python
y.backward()
```
