## **PyTorch AutoGrad**

```python
import torch
```

```python
x = torch.tensor(3., requires_grad=True)
```

```python
y = x**2
```

```python
x 
```

```python
y
```

```python
y.backward()
```

```python
dy_dx = x.grad
```

```python
dy_dx
```

```python
x.grad
```
